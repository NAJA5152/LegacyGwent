```python
import pandas as pd
import shutil
import os
```

```python
# read csv
df = pd.read_csv('audiomap.csv')
```

```python
# clean data
df = df[df['audio0'].notna()]
df = df.iloc[:,1:]
df
```

```python
# transform data
audio0 = df.iloc[:,:3]
audio0 = audio0.rename(columns={'audio0':'audio'})
audio0['suffix'] = '_1'
audio1 = df.iloc[:,:2].join(df['audio1'])
audio1 = audio1.rename(columns={'audio1':'audio'})
audio1['suffix'] = '_2'
audio2 = df.iloc[:,:2].join(df['audio2'])
audio2 = audio2.rename(columns={'audio2':'audio'})
audio2['suffix'] = '_3'
audio3 = df.iloc[:,:2].join(df['audio3'])
audio3 = audio3.rename(columns={'audio3':'audio'})
audio3['suffix'] = '_4'
audio4 = df.iloc[:,:2].join(df['audio4'])
audio4 = audio4.rename(columns={'audio4':'audio'})
audio4['suffix'] = '_5'
audio5 = df.iloc[:,:2].join(df['audio5'])
audio5 = audio5.rename(columns={'audio5':'audio'})
audio5['suffix'] = '_6'
audio6 = df.iloc[:,:2].join(df['audio6'])
audio6 = audio6.rename(columns={'audio6':'audio'})
audio6['suffix'] = '_7'
audio1

audio = pd.concat([audio0 , audio1, audio2, audio3, audio4, audio5, audio6])
audio = audio.dropna()
audio.head()
```
